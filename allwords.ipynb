```python
import pandas as pd


df = pd.read_html('https://www.foregolfleague.com/wordlelist.htm')

```

```python
df[2].fillna("", inplace=True)
df[2]
```

```python
words = []
for i in range(df[2].shape[0]):
    for word in df[2].iloc[i, :].tolist():
        if word == "":
            continue
        words.append(word + '\n')
len(words)

with open('words', 'w') as F:
    F.writelines(words)
```
